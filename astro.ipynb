```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('astro.csv')

ticks = []
for i in range(0, 70, 1):
    ticks.append(i)

plt.figure(figsize=(50,10))
plt.title('Time vs. I', fontsize=50)
plt.xticks(ticks, fontsize=20, rotation=90)
plt.yticks(fontsize=20)
plt.plot(df['time'], df['I'])
plt.show()
```
